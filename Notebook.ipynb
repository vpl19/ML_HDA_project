## First encoding method "One hot encoding"

```python
# class label encoding

df_hot=mushrooms

from sklearn.preprocessing import LabelEncoder
# create instance from LabelEncoder
class_label = LabelEncoder()
# class label to integer
# edible = 0, poisonous = 1
df_hot["class"] = class_label.fit_transform(df_hot["class"].values)
df_hot.head()
```

```python
a=mushrooms.columns
b=a[1:23]
for col in b:
    df_hot = pd.get_dummies(df_hot, columns=[col])
df_hot.head(6)    
```

## Splitting into train and Test 

```python
import numpy as np
from sklearn.model_selection import train_test_split
X=df_hot.loc[:, df_hot.columns != 'class']
y=df_hot["class"]

X_train, X_test, y_train, y_test = train_test_split(
X, y, test_size=0.2, random_state=11)
```

## Supervised Learning 

### KNN

```python
from sklearn.neighbors import KNeighborsClassifier
```

In order to use a KNN we need to encode the data, indeed in order to compute a distance the categorical variables have to be preprocessed 


In order to do that we will use an ordinal encoder https://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.OrdinalEncoder.html

```python
from sklearn.preprocessing import LabelEncoder
labelencoder=LabelEncoder()
for col in mushrooms.columns:
    mushrooms[col] = labelencoder.fit_transform(mushrooms[col])
 
mushrooms.describe()
```
