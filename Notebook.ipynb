## First encoding method "One hot encoding"

```python
# class label encoding

df_hot=mushrooms

from sklearn.preprocessing import LabelEncoder
# create instance from LabelEncoder
class_label = LabelEncoder()
# class label to integer
# edible = 0, poisonous = 1
df_hot["class"] = class_label.fit_transform(df_hot["class"].values)
df_hot.head()
```

```python
a=mushrooms.columns
b=a[1:23]
for col in b:
    df_hot = pd.get_dummies(df_hot, columns=[col])
df_hot.head(6)    
```

## Splitting into train and Test 

```python
import numpy as np
from sklearn.model_selection import train_test_split
X=df_hot.loc[:, df_hot.columns != 'class']
y=df_hot["class"]

X_train, X_test, y_train, y_test = train_test_split(
X, y, test_size=0.4, random_state=11)
```

```python
X_train.describe()
```

```python
y_test.describe()
```

## Supervised Learning 

### KNN

```python
from sklearn.neighbors import KNeighborsClassifier
```

gridsearch to find the best value for K 

```python
from sklearn.model_selection import cross_val_score
import matplotlib.pyplot as plt


```

```python
# search for an optimal value of K for KNN

# list of integers 1 to 15
# integers we want to try
k_range = range(1, 15)

# list of scores from k_range
k_scores = []
k_scores_auc = []

# 1. we will loop through reasonable values of k
for k in k_range:
    # 2. run KNeighborsClassifier with k neighbours
    knn = KNeighborsClassifier(n_neighbors=k)
    # 3. obtain cross_val_score for KNeighborsClassifier with k neighbours
    scores = cross_val_score(knn, X_train, y_train, cv=10, scoring='accuracy')
    
    scores_auc = cross_val_score(knn, X_train, y_train, cv=10, scoring='roc_auc')
    # 4. append mean of scores for k neighbors to k_scores list
    k_scores.append(scores.mean())
    k_scores_auc.append(scores_auc.mean())
#print(k_scores)
```

```python
plt.plot(k_range, k_scores)
plt.xlabel('Value of K for KNN')
plt.ylabel('Cross-Validated Accuracy')
```

```python
# It is always one becuse the class are perfectly separable


plt.plot(k_range, k_scores_auc)
plt.xlabel('Value of K for KNN')
plt.ylabel('Cross-Validated AUC')
```

### Assess the precision of the model 

```python
K=5
# Create KNN classifier
knn = KNeighborsClassifier(n_neighbors = K)
# Fit the classifier to the data
knn.fit(X_train,y_train)
pred=knn.predict_proba(X_test)
pred
```

```python
from sklearn.metrics import roc_curve
fpr, tpr, threshold = roc_curve(y_test, pred[:, 1])
from sklearn.metrics import auc
roc_auc = auc(fpr, tpr)
print(roc_auc)

plt.title('Receiver Operating Characteristic')
plt.plot(fpr, tpr, 'b', label = 'AUC = %0.2f' % roc_auc)
plt.legend(loc = 'lower right')
plt.plot([0, 1], [0, 1],'r--')
plt.xlim([0, 1])
plt.ylim([0, 1])
plt.ylabel('True Positive Rate')
plt.xlabel('False Positive Rate')
```
